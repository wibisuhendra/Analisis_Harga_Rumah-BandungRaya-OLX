```python
from urllib.error import URLError, HTTPError
from urllib.request import urlopen
```

```python
for link in linkarray:
    try:
        page = urlopen('https://www.olx.co.id'+link)
    except (URLError, HTTPError) as e:
        pass
        print("Something bad happened")
    else:
        soup = BeautifulSoup(page.read(),'html.parser')

        adsId.append(soup.find("strong").text)

        title.append(soup.find("h1",{'data-aut-id':"itemTitle"}).text)

        result = soup.find("span",{'data-aut-id':"value_p_bedroom"})
        bedroom.append(result.text if result else "N/A")

        result = soup.find("span",{'data-aut-id':"value_p_bathroom"})
        bathroom.append(result.text if result else "N/A")

        result = soup.find("span",{'data-aut-id':"value_p_sqr_land"})
        landsqr.append(result.text if result else "N/A")

        result = soup.find("span",{'data-aut-id':"value_p_sqr_building"})
        buildingsqr.append(result.text if result else "N/A")

        result = soup.find("span",{'data-aut-id':"value_p_certificate"})
        cert.append(result.text if result else "N/A")

        result = soup.find("span",{'data-aut-id':"value_p_alamat"})
        loc.append(result.text if result else "N/A")

        result = soup.find("span",{'class':"_2FRXm"})
        location.append(result.text if result else "N/A")

        result = soup.find("span",{"data-aut-id":"itemPrice"})
        price.append(result.text if result else "N/A")

        result = soup.find("span",{'data-aut-id':"value_p_floor"})
        floor.append(result.text if result else "N/A")

        result = soup.find("div",{'data-aut-id':"itemCreationDate"})
        postTime.append(result.text if result else "N/A")

        temp = soup.find_all("span",{"class":"_30Ijq"})

        numOfFac.append(len(temp) if temp else "N/A")
        p = ''
        if temp:
            for t in temp:
                p = p + ' ' +t.text+'|'

        fac.append(p if temp else "N/A")

        result = soup.find("div",{'data-aut-id':"itemDescriptionContent"})
        desc.append(result.text if result else "N/A")
#         time.sleep(3)
    
```
